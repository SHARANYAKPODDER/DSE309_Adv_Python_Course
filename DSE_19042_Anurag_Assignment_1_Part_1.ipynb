# Exercise 3: Pig Latin

Pig Latin is a children's secret language. To translate a word into Pig Latin, follow these rules:

1. If the first letter is a vowel (a, e, i, o, or u) then add `way` to the end of the word.
2. Otherwise, move the first letter to the end of the word, and add `ay`.

Some examples:

- `computer` -> `omputercay` 
- `bhopal` -> `hopalbay`
- `table` -> `abletay`
- `apple` -> `appleway`
- `elephant` -> `elephantway`
- `iiser` -> `iiserway`
- `papaya` -> `apayapay`


1. Ask the user to enter a word in English, all lowercase, no punctuation.
2. Print the translation of this word into Pig Latin.

````python
word = input().lower()
vowels = ['a', 'e', 'i', 'o', 'u']
if word[0] not in vowels:
  word = word+word[0]+"ay"
  word = word[1:]
  print(word)
elif word[0] in vowels:
  word = word+"way"
  print(word)
````

# Exercise 5: Restaurant 

1. Define a dict called `menu` in which the keys are the items you can order at the restaurant, and the values are the prices.
2. Define `total` to be 0.
3. Ask the user, repeatedly, to enter what they want to order.
    - If they enter an empty string, stop asking and print the total bill.
    - If they enter an item that *is* on the menu, then print the item, its price, and the new total.
    - If they enter something that is *not* on the menu, then say that you don't have that.
    
Example:

    Order: sandwich
    sandwich is 10, total is 10
    Order: tea
    tea is 5, total is 15
    Order: elephant
    Sorry, we are fresh out of elephant today!
    Order: [ENTER]
    Your total is 15
    
### Hints/notes    

- Define the `menu` dict at the top with keys and values
- Use a `while` loop to get the user's input, since we don't know how many we're going to get.
- Use `in` to check if a key is in the dict

````python
val = True
menu = {"sandwich": 15, "tea": 7, "coffee": 10, "sugar": 2}
total=0
while val:
    order = input("Order:\t").lower()
    if len(order)>0:
      if order in menu.keys():
        total += menu[order]
        print(f"{order} is {menu[order]}, total is {total}")
      else:
        print(f"Sorry, We are out of {order}")
    else:
      val = False
      print(f"Your total is {total}")
````

# Exercise 6: Rainfall

1. Define an empty dict, called `rainfall`.  Eventually, its keys will be strings (names of cities) and its values will be integers (mm of rain).
2. Ask the user repeatedly to enter the name of a city.
3. If they enter an empty string, stop asking and print the dict.
4. If we did get a city, then ask the user to enter mm rain that fell in that city.
5. Accumulate rainfall for this city in the `rainfall` dict.  
   - If this is the first time we're hearing of this city, then just assign the value.
   - If this is not the first time, then add to the previous value
6. Print the dict

Example:

       City: Bhopal
       Rain: 2
       City: Chennai
       Rain: 4
       City: Bhopal
       Rain: 3
       City: [ENTER]
       {'Bhopal':5, 'Chennai':4}

````python
val = True
rainfall = dict()
while val:
    city = input("City:\t")
    if len(city)>0:
      rain_level = int(input("Rain:\t"))
      if city in rainfall.keys():
        rainfall[city] += rain_level
      else:
        rainfall[city] = rain_level
    else:
      val = False
      print(rainfall)
````

# Exercise 7: Simple calculator

1. Define a function, `calc`.
2. In the body of the function:
    - Ask the user to enter a number, and assign it to x.
    - Ask the user to enter a second number, and assign it to y.
    - Ask the user to enter an operator (`+` or `-`), and assign it to `op`
3. Return the result of either adding or subtracting the two numbers.
4. If the user entered a non-number or an operator that we don't recognize, you can return a string with an error... but that's not really the point here.

### Exercise: calc, revisited

Rewrite our `calc` function so that it does *not* use `input` to get the numbers and their operator. Rather, these should be passed as *three* separate arguments, to be placed in parameters `first`, `second`, and `op`.

```python
print(calc(5, 10, '+'))
print(calc(35, 18, '-'))

```

````python
def calc(first, second, op):
  """
  first: first number of int type
  second: second number of int type
  op: either + or -
  """
  try:
    first = float(first)
    second = float(second)
    if op == '+':
      return first+second
    elif op == '-':
      return first-second
    else:
      print("Check op Parameter Type")
      return ValueError
  except ValueError:
    print("Check numeric parameter desc")
    

print(calc(5,8,'-'))
````

# Exercise 8: Biggest and smallest

1. Write a function, `biggest_and_smallest`, that takes a list of numbers.
2. The result from the function will be a two-element tuple containing the smallest number in the input list and the largest number in the input list.
3. Do *not* use any built-in Python functions for sorting, including `min` and `max`.
4. wouldn't it be nice if we could pass our numbers to the function *without* using a list? For this, we have *args

Example:

    biggest_and_smallest([53, 22, 97, 102, 15, 17, 12])
    
That will return the tuple

    (12, 102)

Hints:

- You're going to have to go through each element of the list
- Is it bigger than the biggest thing you've seen so far?  Or is it smaller than the smallest thing you've seen so far?

````python
# With list as parameter
def biggest_and_smallest(list_param):
  """
  list_param: list of numbers
  """
  list_param = list(list_param) # reassures that the param passed is list
  max_num = list_param[0]
  min_num = list_param[0]

  for num in list_param:
    if max_num <= num:
      max_num = num
    if min_num >= num:
      min_num = num
  
  return (min_num, max_num)

biggest_and_smallest([53, 22, 97, 102, 15, 17, 12])
````

````python
# Without passing list
def biggest_and_smallest(*args):
  max_num = args[0]
  min_num = args[0]

  for num in args:
    if max_num <= num:
      max_num = num
    if min_num >= num:
      min_num = num

  return (min_num, max_num)

biggest_and_smallest(53, 22, 97, 102, 15, 17, 12)
````
